# We have 6 Columns that have NULL values

- Director: 2634 null values.
- Cast: 825 null values.
- Country: 831 null values.
- Date added: 10 null values.
- Rating: 4 null values.
- Duration: 3 null values.

```python
# Total NULL values that we have.

print('Total null values :', df.isnull().sum().sum())
```

```python
# Firstly dealing with (Director) column.

df.director = df.director.fillna("Unknown")
```

```python
# Checking "Director" column that is it successfully rectified.

df.director.isnull().sum()
```

```python
#  Now we are dealing with (Cast) column.

df.cast = df.cast.fillna("Not Listed")
```

```python
# Checking "Cast" column that is it successfully rectified.

df.cast.isnull().sum()
```

```python
#  Now we are dealing with (Country) column.

df.country = df.country.fillna("Unknown")
```

```python
# Checking "Country" column that is it successfully rectified.

df.country.isnull().sum()
```

```python
#  Now we are dealing with (Rating) column Bcos after that we drop date_Added and duration  column bcos we have no valueable data in it.

df.rating = df.rating.fillna("Not Listed")
```

```python
# Checking "Country" column that is it successfully rectified.

df.rating.isnull().sum()
```

```python
# Now we drop rest NULL value rows.

df.dropna(inplace=True)
```

```python
# Again checking that now we have NULL values or not.

df.isnull().sum().sum()
```

# Exploratory Data Analysis: Visualization

```python
df
```

```python
# Top Country producing Content.

df.country.value_counts().head(10)
```

```python
sns.barplot(y = df.country.value_counts().head(10).index,x = df.country.value_counts().head(10).values)
plt.xticks(rotation = 90)
plt.title("Visualising Top 10 Countries producing Content")
plt.figure(figsize=(4,3))
```

```python
# Least Countries Producing Content.

df.country.value_counts().tail(10)
```

```python
sns.barplot(y = df.country.value_counts().tail(10).index,x = df.country.value_counts().tail(10).values)
plt.xticks(rotation = 90)
plt.title("Visualising Least 10 Countries producing Content")
plt.figure(figsize=(6,4))
```

```python
# Visualising top 5 Countries producing content.

plt.pie(df.country.value_counts().head(),labels=df.country.value_counts().head().index,startangle=90, autopct='%.1f%%')
plt.legend(loc='upper left', title="Country")
plt.title('Visualising content produced by top 5 countries')
plt.show()
```
